## Methods `2`
### creating custom dataset using Dataset class.
We can add many such attributes and methods similiar to which one are implemented in the std pytroch datasets.

````python
class catdogDataset(Dataset):
    
    def __init__(self, root_dir: str):
                 
        self.root_dir = root_dir
        classes = sorted(entry.name for entry in os.scandir(self.root_dir) if entry.is_dir())
        self.classes = classes

        sub_dirs = glob(self.root_dir + "*")
        samples = [] 
        for dir_ in sub_dirs:
            class_name = dir_.split('/')[-1]
            for img_path in glob(dir_ + '/*'):
                samples.append([img_path, class_name])

                
        class_to_idx = {cls_name: i for i, cls_name in enumerate(classes)}
        targets = [s[1] for s in samples]

        self.class_to_idx = class_to_idx
        self.samples = samples
        self.img_dim = (256,256)
        self.targets = targets
        
    def __len__(self):
        return len(self.samples)
    

    def __getitem__(self, idx):
        img_path, class_name = self.samples[idx]
        with Image.open(img_path) as im:
            im_ = im.resize(self.img_dim)
        transform = transforms.ToTensor()
        img_tensor = transform(im_)
        class_id = self.class_to_idx[class_name]
        return (img_tensor, class_id)
        
        
        
        
        
        
        
````

````python
custom_dataset = catdogDataset(root_dir='./data/test_set/')
````
